## Modelo de oligopolio de Cournot (1838)

Modelo de la interacción entre empresas que compiten por consumidores en cantidades. 

### Modelo general de Cournot

* Un solo bien es producido por $n$ empresas. 
* El costo de la empresa $i$ por producir $q_i$ unidades del bien es $C_i(q_i)$, tal que $C_i$ es una función creciente (a más producción, mayor costo). 
* Toda la producción se vende a un precio único que se determina por la oferta y la demanda de la producción total de la empresa. 
* Sea $Q$ la producción total de la empresa y $P(Q)$ el precio de mercado, en donde $P$ es la función inversa de demanda. 
* Se asume que $P$ es una función decreciente cuando el percio es positivo: si la producción total de la empresa aumenta, entonces el precio disminuye (a menos que el precio sea igual a cero). 
* Si la producción de cada empresa $i$ es $q_i$, entonces el precio es $P(q_1 + \ldots + q_n)$, y los ingresos de la empresa $i$ son $q_i P(q_1 + \ldots + q_n) $. 
* Por tanto, los beneficios de la empresa $i$ son sus ingresos menos costos: 

$$
\pi_i(q_1, \ldots, q_n) = q_i P(q_1 + \ldots + q_n) - C_i(q_i). \tag{56.1}
$$

Cournot fue el primero en sugerir que el comportamiento de la industria se podría caracterizar como un juego estratégico: 

* *Jugadores*: las empresas. 
* *Acciones*: el conjunto de acciones de cada empresa es el conjunto de posibles niveles de producción
* *Preferencias*: las preferencias de cada empresa son representadas por sus beneficios (54.1)

#### Ejemplo: duopolio de Cournot con costos unitarios constantes y función inversa de demanda lineal

* Dos empresas, tal que la función de costos de cada empresa son iguales, $C_i(q_i) = cq_i$ para todo $q_i$, y la función inversa de demanda es lineal y positiva 
$$
P(Q) = 
\begin{cases}
\alpha - Q & \text{ si } Q \leq \alpha \tag{56.2} \\
0 & \text{ si } Q > \alpha
\end{cases}
$$
tal que $\alpha>0$ y $c\geq 0$ son constantes.

* Asuma que $c < \alpha$, por tanto existe un valor de producción total $Q$ tal que el precio de mercado $P(Q)$ es mayor que el costo unitario común $c$. 
Si $c$ fuera mayor que $\alpha$, no podría haber producción de las empresas con beneficios positivos debido a que el precio de mercado no excede el valor de $\alpha$. 

* Para encontrar el equilibrio de Nash, construimos las **funciones de mejor respuesta de las empresas**. 

* Primero se calculan los beneficios de las empresas. 

* Si los productos de las empresas son $q_1$ y $q_2$, el precio de mercado $P(q_1 + q_2)$ es

$$
P(q_1 + q_2) = \begin{cases}
\alpha - q_1 - q_2 & \text{ si } q_1 + q_2 \leq \alpha \\
0 & \text{ si } q_1 + q_2 > \alpha.  \\ 
\end{cases}
$$

* Por tanto, los beneficios de la empresa 1 son

$$
\begin{split}
\pi_1(q_1, q_2) &=& q_1(P(q_1 + q_2) - c) \\ 
 &=& \begin{cases} 
 q_1(\alpha - q_1 - q_2 - c) & \text{ si } q_1 + q_2 \leq \alpha \\
 -cq_1 & \text{ si } q_1 + q_2 > \alpha. \\
 \end{cases}
\end{split}
$$

* Gráfica 57.1 La función inversa de demanda del ejemplo del modelo de Cournot


#### Función de mejor respuesta de producción de la empresa 1 para cualquier valor de producción de la empresa 2: 
 * *primer caso*: si $q_2 = 0$, entonces 
 $\pi_1(q_1,0)= q_1 (\alpha - c - q_1)$ con $q_1 \leq \alpha$ es una función cuadrática. 
 En este caso, 
 $$
 \begin{split}
 \pi_1(0, 0) = 0 \\
 \pi_1(\alpha - c, 0) = 0 
 \end{split}
 $$
y la producción $q_1$ de la empresa 1 maximiza beneficios cuando $q_1 = \frac{1}{2}(\alpha - c)$. 
Por tanto, la función de mejor respuesta de la empresa 1 cuando la producción de la empresa 2 es igual a cero es $b_1(0) = \frac{1}{2}(\alpha - c)$.

* *segund caso*: si $q_2>0$ y $q_2 < \alpha - c$, entonces  
$\pi_1(q_1,q_2)= q_1 (\alpha - c - q_1 - q_2)$ es una función cuadrática. 
En este caso, 
 $$
 \begin{split}
 \pi_1(0, q_2) = 0 \\
 \pi_1(\alpha - c - q_2, q_2) = 0 
 \end{split}
 $$
y la producción que maximiza los beneficios $\pi_1(q_1, q_2)$ es $q_1 = \frac{1}{2}(\alpha - c - q_2)$

* *tercer caso*: si $q_2 > \alpha - c$, entonces $\alpha - c - q_2 <0$. 
Para tales valores de $q_2$, tenemos que 
$\pi_1(q_1,q_2)= q_1 (\alpha - c - q_1 - q_2) <0$ para todos los valores de $q_1$. Por tanto, la mejor respuesta de la empresa 1 es $q_1 = 0$. 

* Se concluye que la mejor respuesta de la empresa 1 ante el nivel de producción de la empresa 2 depende del valor de $q_2$: 
$$
b_1(q_2) = \begin{cases} \frac{1}{2}(\alpha - c -q_2) &\text{ si } q_2 \leq \alpha - c \\
0 &\text{ si } q_2 > \alpha - c
\end{cases}
$$


* Gráfica 58.1 Beneficios de la empresa en función de su producción y dada la producción de la empresas 2


#### Función de mejor respuesta de producción de la empresa 2 para cualquier valor de producción de la empresa 1:
* Debido a que la función de costos de la empresa 2 es la misma que la función de costos de la empresa 1, su función de mejor respuesta de la empresa 2 es la misma: para cualquier valor $q$, tenemos que $b_2(q) = b_1(q)$: 

$$
b_2(q_1) = \begin{cases} \frac{1}{2}(\alpha - c -q_1) &\text{ si } q_1 \leq \alpha - c \\
0 &\text{ si } q_1 > \alpha - c
\end{cases}
$$

* 
$\pi_2(q_1,q_2)= q_2 (\alpha - c - q_1 - q_2)$


#### Funciones de mejor respuesta en el modelo duopólico de Cournot: 
 * cuando la función inversa de demanda está dada por (56.2) y la función de costos de cada empresa es $cq$. 



Un equilibrio de Nash es la dupla de valores de producción $(q_1^*, q_2^*)$ en donde $q_1^*$ es la mejor respuesta ante $q_2^*$, y $q_2^*$ es la mejor respuesta ante $q_1^*$: 
$$
q_1^* = b_1(q_2^*) \text{ y } q_2^* = b_2(q_1^*)
$$
El conjunto de tales duplas de producción es el conjunto de puntos en donde las funciones de mejor respuesta se intersectan en la figura anterior. 

Por tanto, en este caso tenemos dos ecuaciones con dos incognitas: 
$$
\begin{split}
q_1 &=& \frac{1}{2}(\alpha - c - q_2) \\
q_2 &=& \frac{1}{2}(\alpha - c - q_1) 
\end{split}
$$
Resolviendo ambas ecuaciones (sustituyendo la segunda en la primera y despejando $q_1$), encontramos que $q_1^* = q_2^* = \frac{1}{3}(\alpha - c)$

* En resumen, cuando tenemos dos empresas con funciones inversas de demanda $P(Q) = \alpha - Q$ con $Q \leq \alpha$, y el función de costos de cada empresa es $C_i(q_i) = cq_i$, el juego oligopólico de Cournot tiene un equilibrio de Nash único 
$$
(q_1^*, q_2^*) = (\frac{\alpha-c}{3}, \frac{\alpha-c}{3}) 
$$
La producción total en este equilibrio es $\frac{2}{3}(\alpha-c)$, y el precio de equilibrio es $P(\frac{2}{3}(\alpha-c)) = \frac{1}{3}(\alpha+2c)$
Cuando $\alpha$ aumenta (consumidores están dispuestos a pagar más por el bien), el precio de equilibrio y la producción de cada empresa aumenta. 
Cuando $c$ (el costo unitario de producción) aumenta, el producto de cada empresa disminuye y el precio aumenta; cada unidad de incremento en $c$ implica un incremento de dos tercios de la unidad en el incremento del precio. 

## Modelo de oligopolio de Bertrand (1883)

Modelo de la interacción entre empresas que compiten por consumidores en precios. 

### Modelo general de Bertrand

* Un solo bien es producido por $n$ empresas. 
* Cada empresa produce $q_i$ unidades del bien a un costo de $C_i(q_i)$. 
* La función de demanda es $D(p)$
* Asuma que si las empresas fijan diferentes precios, entonces los consumidores compraran el bien de la empresa con el menor precios, ante lo cual la empresa produce la cantidad que sea necesaria para satisfacer la demanda. 
* Si más de una empresa fijan el menor precio, todas estas empresas comparten la demanda a ese precio en forma igualitaria. 
* La empresa que no fije el menor precio no recibe demanda y por tanto no produce. 

En resumen, **el juego oligopólico de Bertrand** se caracteriza por el siguiente juego de estrategias: 

* *jugadores*: las empresas 
* *acciones*: el conjunto de acciones de las empresas es el conjunto de posibles precios (nonegativos)
* *preferencias*: las preferencias de la empresa $i$ son representadas por sus beneficios, igual a  $p_i D(p_i)/m - C_i(D(p_i)/m)$ si la empresa $i$ es una de las $m$ empresas que fijaron el menor precio , y beneficios igual a cero si alguna fija un precio menor a $p_i$. 

#### Ejemplo: duopolio de Bertrand con costos unitarios constantes y función inversa de demanda lineal

* Dos empresas con costos unitarios constante $c$, esto es, la función de costos de cada empresa son iguales, $C_i(q_i) = cq_i$, con $i=1,2$
* Suponga que la función de demanda es $D(p) = \alpha - p$ para $p\leq \alpha$ y $D(p) = 0$ si $p>\alpha$, y $c<\alpha$. 

* debido a que el costo unitario de producción es el mismo e igual a $c$, la empresa $i$ genera beneficios igual a $p_i - c$ por cada unidad que vente: 
$$
\pi_i(p_1, p_2) = \begin{cases}
(p_i-c)(\alpha-p_i) \text{ si } p_i < p_j \\
\frac{1}{2}(p_i - c)(\alpha-p_i) \text{ si } p_i = p_j \\
0 \text{ si } p_i > p_j,
\end{cases}
$$
donde $j$ es la otra firma ($j=2$ si $i=1$, y $j=1$ si $i=2$). 

* Para encontrar el equilibrio de Nash del juego se requiere encontrar las funciones de mejor respuesta de las empresas.
